```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
x = np.array([12,23,34,45,49,56,61,68,71,78])

y = np.array([23,31,38,43,49,56,65,71,79,85])

z = np.array([98,89,86,76,71,68,61,54,49,34])
```

### Line Plot

```python
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x, y, marker = 'D', color = 'b', mfc = 'r', ls = '--', ms = 7)
```

```python
plt.title('X vs Z')
plt.xlabel('First Value')
plt.ylabel('Third Value')
plt.plot(x, z, marker = 'p', color = 'maroon', mfc = 'yellow', ls = '-.', ms = 9)

plt.grid()
```

```python
plt.title('X vs Z')
plt.xlabel('First Value')
plt.ylabel('Second and Third Value')
plt.plot(x, z, marker = 'p', color = 'maroon', mfc = 'yellow', ls = '-.', ms = 9)
plt.plot(x, y, marker = 'D', color = 'blue', mfc = 'r', ls = '--', ms = 7)
plt.grid()
```

### Subplot

```python
plt.figure(figsize=(10,8))

plt.subplot(2,2,1)

plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x, y, marker = 'D', color = 'blue', mfc = 'r', ls = '--', ms = 7)

plt.grid()
plt.subplot(2,2,2)

plt.title('X vs Z')
plt.xlabel('First Value')
plt.ylabel('Third Value')
plt.plot(x, z, marker = 'p', color = 'maroon', mfc = 'yellow', ls = '-.', ms = 9)

plt.grid()
```

### Bar Graph

```python
students = [56, 65, 46, 51, 61]

courses = ['Java', 'Python', 'R', 'Kotlin', 'Dart']
```

```python
plt.title('Course Count')
plt.xlabel('Courses')
plt.ylabel('Students')
plt.bar(courses, students, color='y', edgecolor = 'black', 
        width = 0.5, linewidth = 2, alpha = 0.8)
```

### Scatter Plot

```python
plt.title('Scatter Plot')
plt.scatter(x, y, marker = 'o')
plt.xlabel('First Values')
plt.ylabel('Second Values')
plt.grid()
```

### Pie Chart

```python
palette = ['indigo','purple','b','skyblue','lightblue']
plt.title('Pie Chart')
plt.pie(students, labels=courses, autopct = '%.2f%%',
        startangle=180, radius = 1.25, labeldistance= 1.05, colors = palette,
        explode = [0,0,0.15,0,0])
```

```python
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)

plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x, y, marker = 'D', color = 'blue', mfc = 'r', ls = '--', ms = 7)


plt.subplot(2,2,2)

palette = ['indigo','purple','b','skyblue','lightblue']
plt.title('Pie Chart')
plt.pie(students, labels=courses, autopct = '%.2f%%',
        startangle=180, radius = 1.25, labeldistance= 1.05, colors = palette,
        explode = [0,0,0.15,0,0])

plt.subplot(2,2,3)

plt.title('Bar Graph')
plt.xlabel('Courses')
plt.ylabel('Students')
plt.bar(courses, students, color='y', edgecolor = 'black', 
        width = 0.5, linewidth = 2, alpha = 0.8)

plt.subplot(2,2,4)

plt.title('Sactter Plot')
plt.xlabel('First Values')
plt.ylabel('Second Values')
plt.scatter(x, y, marker = '^')
plt.grid()


plt.savefig('My Plots')
```

### Box Plot

```python
marks = [67, 63, 70, 71, 72, 56, 51, 21, 95]
```

```python
plt.boxplot(marks)
```

```python
q3 = np.percentile(marks, 75)

q1 = np.percentile(marks, 25)
```

```python
q3, q1
```

```python
iqr = q3 - q1
iqr
```

```python
# Lower bound
lb = q1 - 1.5 * iqr

# Upper bound
ub = q3 + 1.5 * iqr
```

```python
lb, ub
```

### Seaborn

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault)
```

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = tips, hue = 'time')
plt.show()
```

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = tips, hue = 'sex')
plt.show()
```

### Histogram

```python
plt.xlabel('Tip Amount')
plt.ylabel('Frequency')
sns.histplot(tips['tip'], kde = True)
plt.show()
```

```python
sns.boxplot(tips['tip'])
plt.show()
```

```python
plt.title('Outliers in Tip by Days')
plt.xlabel('Day')
plt.ylabel('tip')
sns.boxenplot(x = 'day', y = 'tip', data = tips)
plt.show()
```

```python
sns.pairplot(tips, hue = 'smoker')
plt.show()
```
