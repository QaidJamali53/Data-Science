#### Standard Deviation

```python
x = np.array([67,54,17,83,14,59,63,77])
```

```python
plt.scatter(x, range(8))
```

```python
x.mean()
```

```python
plt.scatter(x, range(8))
plt.scatter(x.mean(), 3.5, color='r')
```

```python
std = np.sqrt(np.sum((x- x.mean())**2)/len(x))
```

```python
std
```

```python
x.std()
```

#### Standardization    

```python
# It can make the standard deviation = 1 for dataset
def standardize(x):
    val = (x - x.mean())/ x.std()
    return val
```

```python
x_new = standardize(x)
```

```python
x_new.std()
```

```python
plt.subplot(1,2,1)
plt.scatter(x, range(8))
plt.subplot(1,2,2)
plt.scatter(x_new, range(8), color='r')
```

```python
df = pd.read_csv('wisc_bc_data.csv')
```

```python
df
```

```python
# Seperate Input and Output Variables
x = df.drop(['id', 'diagnosis'], axis=1)
y = df['diagnosis']
```

```python
x.shape
```

```python
from sklearn.feature_selection import SelectKBest, chi2
```

```python
skf = SelectKBest(score_func = chi2, k=5)
```

```python
skf.fit(x,y)
```

```python
skf.get_support()
```

```python
x_new = x.iloc[:, skf.get_support()]
x_new
```
